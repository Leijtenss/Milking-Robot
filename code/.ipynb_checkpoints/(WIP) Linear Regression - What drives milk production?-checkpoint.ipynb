## What drives milk production?
In this part, we're going to load the following tables and explore them:
<br>
* HistoryAnimal: acts as a DIM-table for the cows in the livestock, containing variables such birth data, number of calves, and breed (BasicAnimal would be an alternative to this table)
* HistoryAnimalDailyData: a table with milking data per cow. Even more data is present in this table compared to AnimalDaily, over 180k records
* SessionMilkYield: table containing information on specific milk sessions. Can be used to calculate time of day of sessions, duration of sessions etc. Has data from March 2017 to April 2018 so might be a good period to start with
* AnimalFeedDailyConsumption: contains information on how much cows have been fed, besides by grazing outside, which cannot be measured with the milking robot. Contains data for a full year, April 2017 - April 2018, which is the time period we'll focus on initially. 
<br>
<br>
After exploring, we're going to try and estimate our first linear regression and see what characteristics (cow, session and feed) affect the total yield (in liters of milk) they provide.

```python
## Load the tables from the database
## We'll end up with four dataframes: dim_cow, daily, session and feed
# HistoryAnimal
sql_dim = '''
SELECT *
FROM dbo.HistoryAnimal;
'''
dim_cow = pd.read_sql_query(sql_dim, db)

# HistoryAnimalDailyData
sql_daily = '''
SELECT *
FROM dbo.HistoryAnimalDailyData
WHERE DayDate >= '2017-04-10';
'''
daily = pd.read_sql_query(sql_daily, db)

# SessionMilkYield
sql_session = '''
SELECT *
FROM dbo.SessionMilkYield
WHERE BeginTime >= '2017-04-10';
'''
session = pd.read_sql_query(sql_session, db)

# AnimalFeedDailyConsumption
sql_feed = '''
SELECT *
FROM dbo.AnimalFeedDailyConsumption
WHERE DateAndTime >= '2017-04-10';
'''
feed = pd.read_sql_query(sql_feed, db)
```

#### Create extra variables

```python
session.sort_values(['BasicAnimal', 'BeginTime']).head()
```

```python
# Replace missing PreviousEndTime with EndTime if missing
session['PreviousEndTime'] = np.where(session['PreviousEndTime'].isnull(), session['EndTime'], session['PreviousEndTime'])
```

```python
# Create Yield-variables
session['DiffYield'] = session['TotalYield'] - session['ExpectedYield']
session['DiffYield%'] = session['DiffYield'] / session['ExpectedYield']
```

```python
## Create several time variables
# Date
session['Date'] = session['BeginTime'].dt.date

# SessionTime
session['SessionTime'] = session['EndTime'] - session['BeginTime']

# TimeSincePreviousSess
session['TimeSincePreviousSess'] = session['BeginTime'] - session['PreviousEndTime']
```

```python
# Create session_agg
session_agg = session.groupby(['BasicAnimal', 
                               'AnimalDaily', 
                               'Date']).agg({'SessionNo':['max'],
                                             'TotalYield':['min', 'max', 'sum'],
                                             'ExpectedYield':['min', 'max', 'sum'],
                                             'DiffYield':['min', 'max', 'sum'],
                                             'SessionTime':[np.min, np.max, np.sum],
                                             'TimeSincePreviousSess':[np.min, np.max, np.sum],
                                             'AvgConductivity':['mean']
                                            }).reset_index()

session_agg.columns = ["_".join(x) for x in session_agg.columns.ravel()]

session_agg.iloc[:,3:] = session_agg.iloc[:,3:].round(2)

session_agg = session_agg.rename(columns=({'BasicAnimal_': 'BasicAnimal',
                                          'AnimalDaily_': 'AnimalDaily',
                                          'Date_': 'Date'}))
```

```python
import matplotlib.dates as mdates
labels = ['apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec', 'jan', 'feb', 'mar', 'apr']

# Expected Yield vs Total Yield over time
plt.subplots(nrows=1, ncols=2, figsize=(20,8))

plt.subplot(1,2,1)
plt.title('TotalYield vs Expected Yield over time')
ax = session['TotalYield'].groupby(session['BeginTime'].dt.month).sum().plot()
ax = session['ExpectedYield'].groupby(session['BeginTime'].dt.month).sum().plot()
ax.legend(["TotalYield", "ExpectedYield"])
ax.set_xlabel('Months')

plt.subplot(1,2,2)
plt.title('Deviation over time')
ax1 = session['DiffYield'].groupby(session['BeginTime'].dt.month).sum().plot.bar(color='slategrey')
ax1.legend(['DiffYield'])

plt.tight_layout()
```

### Imports & connecting to the SQL-database

## Start with exploring AnimalDaily
Looks like this table, one containing data on the daily milk yield per cow, is a good one to start with.
<br>
Let's import it into dataframe as df_ad

```python
sql_ad = '''
SELECT *
FROM dbo.AnimalDaily;
'''

df_ad = pd.read_sql_query(sql_ad, db)
df_ad.head(5)
```

```python
#Let's get some basic statistics to start with
df_ad.info()
df_ad.describe()
```

```python
# We see that Date is stored as an object, convert to date(time)
df_ad['Date'] = pd.to_datetime(df_ad['Date'])
```

## Time-span: what time period are we talking about here?
Let's do a count per date and see how many cows have been milked by the robot over time, in order to get a sense of how good our data is and how large the herd is we're looking at.

```python
# Check first and last date in dataset
print("First date: " + str(df_ad['Date'].min()))
print("Last date: " + str(df_ad['Date'].max()))
print("Number of rows in dataset: " + str(len(df_ad.index)))
```

Seems like quite a long period of data we have in our database, over 33k records ranging from 2005 to 2018. <br>
Before moving on to checking out the number of cows and milk over time, let's inspect the quality of our data and see if we have any missing values.

## Let's check for missing values
It is important to see if we're missing values for for instance TotalYield and Duration, since these are some of the most important fields to visualize later on.


```python
df_ad.isnull().sum()
```

__Note__ there are 34 missings for AnimalGroup, which should be fine. <br>
As we saw in the general descriptive statistics, most cows belong to group 3 so there is not much variation there. <br>
<br>
There are 499 missing rows however for TotalYield and Duration, which might be a bit more troublesome. <br>
Let's focus on these rows first and see if they are problematic or not. This allows to determine our course of action on what to do with them.

```python
# Select the portion of dataframe with missing values for the column "TotalYield"
df_missing = df_ad[pd.isnull(df_ad["TotalYield"])]
df_missing.head(5)
```

It looks like that for some cows TotalYield is missing because DSLC _or days since last calf_ is 0.<br>
However, this does not hold true for all, shown by OID 1401 and 2132 for instance where LactationNumber is 4 and 1 respectively. <br>
<br>
What to do now? First, let's check out how many rows with missing TotalYield we have per value of LactationNumber (or number of calves a cow has had).<br>
<br>
It is good to know that a cow cannot start to produce milk without having given birth to at least one calf.

```python
df_missing.groupby('LactationNumber').agg('count').OID
```

As it turns out we see that the majority of rows (305 out of 499) don't have a TotalYield because the cow didn't give birth to a calf yet. We'll filter out these rows from the dataset.
<br>
This leaves us with 194 rows which we'll filter out as well for the time being. 

```python
del df_missing
df_ad = df_ad.dropna(subset = ['TotalYield'])
```

## Continue with exploring the data

Let's start with checking out the amount of records over time

```python
df_ad = df_ad.sort_values('Date')
```

Before plotting the count of records over time, let's increase default size of the plot

```python
df_ad['Date'].value_counts().plot(figsize = (12.5,7.5))
```

As we can see there are not many datapoints in the beginning, it seems that things are taking of from 2017 onwards <br>
Also, the plot is a bit small, let's see if we can make subplots per year and/or month that are a bit more clear.

```python
# Add Year, Month, Day, and Yrmnth
df_ad['YearMonth'] = df_ad['Date'].map(lambda x: 100*x.year + x.month)
df_ad['Year'] = df_ad['Date'].dt.year
df_ad['Month'] = df_ad['Date'].dt.month
df_ad['Day'] = df_ad['Date'].dt.day
```

Check if we can plot trends for number of records, TotalYield and Duration per YearMonth, side by side

```python
# Aggregate the df and check for # of records, # of unique cows, and sum of TotalYield and Duration
df_agg = df_ad.groupby(['YearMonth']).agg({'Year': max,
                                           'Month': max,
                                           'OID': "count", 
                                           'BasicAnimal': pd.Series.nunique, 
                                           'TotalYield': sum, 
                                           'Duration': sum})
df_agg = df_agg.reset_index()
df_agg.head(5)
```

```python
# Add some extra columns, such as YieldPerCow, YieldPerTurn, and DurationPerTurn
# And then filter out all years before 2015
df_agg['YieldPerCow'] = df_agg['TotalYield'] / df_agg['BasicAnimal']
df_agg['YieldPerTurn'] = df_agg['TotalYield'] / df_agg['OID']
df_agg['DurationPerTurn'] = df_agg['Duration'] / df_agg['OID']
df_agg = df_agg[df_agg['Year']>=2015]
```

Now we have some decent summary stats that we can visualize <br>
Start with looking at the number of cows milked over the years

```python
g = sns.catplot("Month","BasicAnimal", 
            data=df_agg, 
            kind="bar",
            col='Year',
            palette="Blues",
            height=4,
            aspect=1
           )

g.set_ylabels('Number of Cows')

# Nog aanpassen naar multiple subplots, 4 in een quadrant
# Misschien ook nog andere kleuren, maar dat is een latere zorg
```

It looks like the milking robot started to being used more consistently from 2015, stabilizing from the November 2016 onwards.
<br>
It is interesting to see that we see an unusual spike in April 2017 (or month 4 in 2017). For the other months of the year we see that the livestock consists of rougly 70 cows, whereas it peaks in April with over a 100 cows.
<br>
<br>
Let's zoom in on the first couple of months of 2017 in order to find out what drives this sudden increase.
<br>

```python
# Let's zoom in on February through June 2017 and get some basic stats
filter_list = ('201702', '201703', '201704', '201705', '201706')
df_agg[(df_agg.YearMonth.isin(filter_list))]
```

Looking at TotalYield and YieldPerCow we immediately see something interesting for April: <br>
TotalYield is similar to the previous month and the one after that, but YieldPerCow is at least 200 liters less.

```python
# Not getting anywhere with data on a monthly basis, let's look at the daily level
# Aggregate the df and check for # of records, # of unique cows, and sum of TotalYield and Duration
df_agg = df_ad.groupby(['Date']).agg({'YearMonth': max,
                                      'Year': max,
                                      'Month': max,
                                      'Day': max,
                                      'OID': "count", 
                                      'BasicAnimal': pd.Series.nunique, 
                                      'TotalYield': sum, 
                                      'Duration': sum,
                                      'IsYieldValid': sum
                                     })

df_agg['YieldPerCow'] = df_agg['TotalYield'] / df_agg['BasicAnimal']
df_agg = df_agg.reset_index()

df_agg = df_agg.copy()
df_agg['IsYieldValid'] = df_agg['IsYieldValid'].astype('int64', copy=False)

# Create subset for 3 months in 2017
months = (201703, 201704, 201705)

df_agg_17 = df_agg[(df_agg.YearMonth.isin(months))]
```

Let's see how the unique number of cows milked per day per month compares to the average yield per day. <br>

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12,10))

sns.lineplot(x='Day', 
            y='BasicAnimal', 
            data=df_agg_17, 
            hue='YearMonth',
            palette="Blues",
            legend='full',
            ax=ax[0]
           )

sns.lineplot(x='Day', 
            y='YieldPerCow', 
            data=df_agg_17, 
            hue='YearMonth',
            palette="Blues",
            legend='full',
            ax=ax[1]
           )

plt.show()
```

A quick look at the first 5 days of 201704 shows us that there was an extreme high number of unique cows being milked, but this does not result in a average YieldPerCow. Therefore we need to go deeper into the details of the first week of 201704 and see what drives this.

```python
df_ad[df_ad['Date'] == '2017-04-03'].head(10)
```

First thing that caught my eye is the large number of rows with 0 TotalYield and IsYieldValid 0 too. <br>
Plot the number 1's and 0's for IsYieldValid first, then filter out if needed. <br>
<br>
Note: it would be interesting though to look at these erroneous records later on and see how they behave over time. Let's start with Mar-May 2017.

```python
# First, filter on months and filter out valid records
subset = df_ad[(df_ad.YearMonth.isin(months)) & (df_ad['IsYieldValid'] == 0)]

sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(12.5,7.5))

# Define plot
ax = sns.countplot(x='IsYieldValid', 
                   hue="YearMonth",
                   data=subset)           

plt.show()
```

As you can see, a clearly there is a higher number of invalid yield for that month. This means that we've been looking at our data in the wrong way and we might need to filter out all records with an invalid yield. <br>
<br>
Before doing that, let's calculate the share of IsYieldInvalid per month (>= 2015) to get an idea what it looks like.

```python
df_agg_15 = df_agg[df_agg['Year']>=2015]

df_agg_15 = df_agg_15.copy()
df_agg_15['ErrorRate'] = 1 - (df_agg_15['IsYieldValid'] / df_agg_15['OID'])
```

```python
df_agg_15.head(10)
```

```python
# Plot ErrorRate over time
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(12.5,7.5))

# Define plot
sns.lineplot(x='Month', 
            y='ErrorRate', 
            data=df_agg_15, 
            hue="Year",
            palette="Blues",
            legend='full',
            ax=ax
           )

plt.show()
```

It seems that the average rate of invalid milk yield was quite high during 2015 and 2016, stabilizing throughout 2017.

Now, continue to filtering out the erroneous records and visualizing the unique number of cows milked over time

```python
df_ad_1 = df_ad[df_ad['IsYieldValid'] == 1]
```

Let's see how the pattern over time looks now with all the rows with invalid yield filtered out

```python
# First, recreate initial df_agg and filter out everything after 2015
df_agg = df_ad_1.groupby(['Date']).agg({'YearMonth': max,
                                      'Year': max,
                                      'Month': max,
                                      'Day': max,
                                      'OID': "count", 
                                      'BasicAnimal': pd.Series.nunique, 
                                      'TotalYield': sum, 
                                      'Duration': sum})

df_agg = df_agg.reset_index()
df_agg = df_agg[df_agg['Year']>=2015]
```

```python
# Then create catplot
g = sns.catplot("Month","BasicAnimal", 
            data=df_agg, 
            kind="bar",
            col='Year',
            palette="Blues",
            height=4,
            aspect=1
           )

g.set_ylabels('Number of Cows')
```

This clearly shows that we need to be aware of taking into account IsYieldValid when analyzing Yield.
